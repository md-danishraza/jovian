#### 19. Create a 8x8 matrix and fill it with a checkerboard pattern (★☆☆)

````python
arr = np.zeros(shape=(4,4),dtype=np.int8)

for index,values in np.ndenumerate(arr):
    if (index[0]+index[1])%2==0:
        arr[index[0],index[1]] = 1

print(arr)

````

#### 23. Create a custom dtype that describes a color as four unsigned bytes (RGBA) (★☆☆)

````python
#defining custom datatype for rgba
#each color is tuple for rgba numbers
rgba_dtype = np.dtype([
    ('red', np.uint8),
    ('green', np.uint8),
    ('blue', np.uint8),
    ('alpha', np.uint8)
])

#creating array of colors
colors = np.array([
    (255, 0, 0, 255),     # Red with full opacity
    (0, 255, 0, 255),     # Green with full opacity
    (0, 0, 255, 255),     # Blue with full opacity
    (255, 255, 0, 128),    # Yellow with 50% opacity
    (60, 24, 218, 0.8*255)     #kind of purple
], dtype=rgba_dtype)

print(colors['blue'])   #all values of blues channels
print(colors[2])      #third color rgba
````

#### 25. Given a 1D array, negate all elements which are between 3 and 8, in place. (★☆☆)

````python
arr = np.array([2,3,7,5,6,10,13])

#we have to inclusively negate all the elements between 3 and 8
for index,value in np.ndenumerate(arr):
  if 3<value<8:
    arr[index[0]] = np.negative(value)

print(arr)
````

#### 57. How to randomly place p elements in a 2D array? (★★☆)

````python
main = np.arange(10,20).reshape(10)
sample = np.array([1,2,3,4,5])

#random index for sample array elements
indices = np.random.choice(main.size,sample.size,replace=False)
print(indices)

#replacing sample values as random index
for idx,value in zip(np.nditer(indices),np.nditer(sample)):
    main[idx] = value

print(main)
````

#### 58. Subtract the mean of each row of a matrix (★★☆)

````python

mat = np.arange(20).reshape(4,5)
print(mat)
#row wise sum horizontal 1
rowsum = mat.sum(1)
print(rowsum)
#mean of rows
rowmean = rowsum/(mat.shape[1])

#reshaping then transposing into 4*1
rowmean = np.reshape(rowmean,(1,4)).T
#print(rowmean.shape)

#subract mean of each row
print(mat-rowmean)
````

#### 60. How to tell if a given 2D array has null columns? (★★☆)

````python
arr = np.array([[1, 0, 3, 0],
                [0, 0, 0, 0],
                [5, 0, 0, 0]])
#row by row sum hence vertical
colsum = arr.sum(axis=0)
print(colsum)

print("folowing nth columns are null :-")
for idx,sum in enumerate(colsum):
    if sum == 0:
        print(idx,end=",")
````
